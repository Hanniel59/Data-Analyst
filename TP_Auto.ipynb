##Répartition des vehicules par année d'immatriculation

```python
df3=df2[df2['yearOfRegistration'] >1960]
```

```python
annee_imm=df3['yearOfRegistration'].value_counts().sort_index(ascending=True)
```

```python
plt.figure(figsize=(12, 6))

# Utiliser une palette de couleurs plus attrayante et ajouter des étiquettes
plot = annee_imm.plot(kind='bar', color='#6CAFB7')

plt.title("Nombre total de véhicules par année d'immatriculation", fontsize=16)
plt.xlabel('Année d\'immatriculation', fontsize=12)
plt.ylabel('Nombre total de véhicules', fontsize=12)
plt.xticks(rotation=45, ha='right', fontsize=8)
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Ajouter une grille en arrière-plan
plt.tight_layout()

plt.show()
```

## Prix moyen par type de véhicule et boite

```python
prix_moyen= df2.groupby(['vehicleType', 'gearbox'])['price'].mean().reset_index()
```

```python
prix_moyen
```

```python


plt.figure(figsize=(8, 6))
heatmap_data = prix_moyen.pivot('vehicleType', 'gearbox', 'price')

# Détermination des limites pour la colorbar
vmin = heatmap_data.min().min()
vmax = heatmap_data.max().max()

# Création du heatmap avec la palette de couleurs et définissant les limites
sns.heatmap(heatmap_data, annot=True, cmap='YlGnBu', fmt='.2f', vmin=vmin, vmax=vmax)

plt.title('Heatmap des données de véhicule par type de véhicule et boîte de vitesses')
plt.xlabel('Type de boite de vitesse')
plt.ylabel('Type de véhicule')

# Agrandir la taille des annotations pour une meilleure lisibilité
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)


plt.tight_layout()
plt.show()
```
